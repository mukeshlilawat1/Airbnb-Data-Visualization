```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Airbnb_Data.csv")
```

```python
df.head()
```

```python
df.columns
```

# checking missing value

```python
print(df.isnull().sum())
```

```python
df.info()
```

# handling missing values

```python
df['last_review'] = pd.to_datetime(df['last_review'], errors= 'coerce')
```

```python
df.info()
```

```python
df.fillna({'review_scores_rating' : 0, 'last_review' : df['last_review'].min()}, inplace=True)
```

```python
df.dropna(subset= ['name', 'neighbourhood'], inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns=["description", "thumbnail_url"], errors='ignore')
```

```python
df.head()
```

## Remove Duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

# descriptive statistics

```python
df.describe()
```

# visualization

```python
plt.figure(figsize=(10,6))
sns.histplot(df['log_price'], bins=50, kde=True, color='blue')
plt.title("Distribution of listing price")
plt.xlabel("log_price $")
plt.ylabel("frequency")
plt.show()
```
